```python
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

```

```python
#example 1

data, true_labels = make_blobs(n_samples=1000, centers=3, cluster_std=2)
plt.scatter(data[:, 0], data[:, 1], c=true_labels, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Generated Data')
plt.show()
```

```python
gmm1=GaussianMixture(n_components=3,random_state = 42)
gmm1.fit(data)

pred_labels = gmm1.predict(data)
cluster_centers = gmm1.means_


#plot the prediction

# Scatter plot of the data points, colored by predicted labels
plt.scatter(data[:, 0], data[:, 1], c=pred_labels, cmap='viridis', label='Data Points')

# Plot the cluster centers
for idx,(x,y) in enumerate(cluster_centers):
  plt.text(x,y,str(idx), color= "black", fontsize = 12, ha = "center", va = "center",
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.2'))

# Add labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('GMM Clustering Results')

# Show the plot
plt.show()
```
